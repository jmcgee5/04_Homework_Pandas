```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#for col in school_data_complete.columns: 
 #   print(col)     
    
Total_Schools = len(school_data['School ID'].unique())
#print (Total_Schools)

Total_Students = len(student_data['Student ID'].unique())
#print (Total_Students)

Total_Budget = school_data['budget'].sum()
#print (Total_Budget)

Avg_Math = student_data['math_score'].mean()
#print (Avg_Math)

Avg_Read = student_data['reading_score'].mean()
#print (Avg_Read)

Overall_Passing_Rate = (Avg_Math + Avg_Read)/2
#print (Overall_Passing_Rate)

Pct_Passing_Math = (len(student_data.loc[student_data['math_score']>=70]))/Total_Students
#print (Pct_Passing_Math)

Pct_Passing_Reading = (len(student_data.loc[student_data['reading_score']>=70]))/Total_Students
#print (Pct_Passing_Reading)

pcs_df = pd.DataFrame({
    'Total Schools': [Total_Schools]
    , 'Total Students': [Total_Students]
    , 'Total Budget': [Total_Budget]
    , 'Avg Math Score': [Avg_Math]
    , 'Avg Reading Score': [Avg_Read]
    , '% of Passing Math': [Pct_Passing_Math]
    , '% of Passing Reading': [Pct_Passing_Reading]
    , 'Overall Passing Rate': [Overall_Passing_Rate]
    
})

pcs_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  

* Create a dataframe to hold the above results

```python
#Add column for Passing Math students and show as True/False
school_data_complete['Passing_Math'] = school_data_complete['math_score'] >= 70

#Add column for Passing Reading students and show as True/False
school_data_complete['Passing_Reading'] = school_data_complete['reading_score'] >= 70

#Add column for Passing Overall students and show as True/False
school_data_complete['Passing_Overall'] = school_data_complete['Passing_Math'] & school_data_complete['Passing_Reading']

#remove comment to show new columns added
#school_data_complete

#group by school id
gb_schools_stu = school_data_complete.groupby('School ID')

#use funcs to name a column and the type of aggregation
funcs = {'Student ID': 'count', 'math_score': 'mean', 'reading_score': 'mean', 'Passing_Math': 'sum', 'Passing_Reading': 'sum', 'Passing_Overall': 'sum'}

#create aggregated table with agg command
school_agg = gb_schools_stu.agg(funcs)

#remove comment to show aggregates
#school_agg

#merge the schools data with the aggregated data to create a new table
school_sum = pd.merge(school_data[['school_name','type','budget']], school_agg, how="left", left_index=True, right_index=True)

#Add column for Per Student Budget
school_sum['Per Student Budget'] = school_sum['budget'] / school_sum['Student ID']

#Add column for pct Passing Math
school_sum['% Passing Math'] = school_sum['Passing_Math'] / school_sum['Student ID']

#Add column for pct Passing Reading
school_sum['% Passing Reading'] = school_sum['Passing_Reading'] / school_sum['Student ID']

#Add column for pct Passing Overall
school_sum['% Passing Overall'] = school_sum['Passing_Overall'] / school_sum['Student ID']

school_sum

```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_5 = school_sum.sort_values(by='% Passing Overall', ascending=False).head(5)
top_5



```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_5 = school_sum.sort_values(by='% Passing Overall', ascending=False).tail(5)
bottom_5
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Set up bins for spending ranges with a list
spending_range_bins = [0, 585, 615, 645, 675]

#Give the rows for the Budget Per Student Range
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#Use pd.cut to assign the spending cell value 
spending = pd.cut(school_sum['Per Student Budget'], spending_range_bins, labels=group_names)

#show result of cutting into bins
#spending

#Add column with bins
school_sum['Budget per Student Range'] = spending

#Group by Budget
budget_groups = school_sum.groupby(by='Budget per Student Range')

budget_groups['math_score', 'reading_score', '% Passing Math', '% Passing Reading', '% Passing Overall'].mean()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Set up bins for school sizes ranges with a list
size_bins = [0, 1000, 2000, 5000]

#Give the rows for the School Size
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#Use pd.cut to assign the spending cell value
size = pd.cut(school_sum['Student ID'], size_bins, labels=group_names)

#show result of cutting into bins
#size

#Add column with size
school_sum['School Size'] = size

#Group by School Size
size = school_sum.groupby(by='School Size')

size ['math_score', 'reading_score', '% Passing Math', '% Passing Reading', '% Passing Overall'].mean()
```
